# Load Data

```python
import pandas as pd

# Load data
# Implement me
df = pd.read_csv('https://archive.ics.uci.edu/ml/machine-learning-databases/hepatitis/hepatitis.data',
                 header=None)

# Specify the name of the columns
df.columns = ['Target', 'AGE', 'SEX', 'STEROID', 'ANTIVIRALS', 'FATIGUE', 'MALAISE', 'ANOREXIA', 'LIVER BIG', 'LIVER FIRM', 'SPLEEN PALPABLE', 'SPIDERS', 'ASCITES', 'VARICES', 'BILIRUBIN', 'ALK PHOSPHATE', 'SGOT', 'ALBUMIN', 'PROTIME', 'HISTOLOGY']

# Specify the name of the target
target = 'Target'

df.head()
```

# Check missing values

```python
import numpy as np

# Replace ? with np.NaN
# Implement me
df = df.replace('?', np.NaN)

# Check NA
df.isna().sum()
```

# Remove rows with missing values

```python
print('Number of rows before removing rows with missing values:', str(df.shape[0]))

# Remove rows with np.NaN
# Implement me
df = df.dropna(how='any')

print('Number of rows after removing rows with missing values:', str(df.shape[0]))
```

# Get the features and target

```python
# Implement me
X = df.drop(columns=[target])
y = df[target]
```

# Check categorical features

```python
# Print the unique value and their number for each feature
for j in range(X.shape[1]):
    print(X.columns[j] + ':')
    print(X.iloc[:, j].value_counts(), end='\n\n')
```

# Check categorical target

```python
# Print the unique value and their number for the target
y.value_counts()
```

# Over sampling

```python
from imblearn.over_sampling import RandomOverSampler

# RandomOverSampler (with random_state=0)
# Implement me
ros = RandomOverSampler(random_state=0)
X, y = ros.fit_sample(X, y)

pd.DataFrame(data=y, columns=[target])[target].value_counts()
```

# A brute-force method for hyperparameter tuning
- Here we use a nested for-loop to fine-tune three hyperparameters of SVC
    - C (penalty parameter of the error term)
    - kernel
    - gamma (kernel coefficient)

```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.svm import SVC
from sklearn.model_selection import StratifiedKFold
from sklearn.model_selection import cross_val_score

# The list of value for hyperparameter C (penalty parameter of the error term)
Cs = [0.01, 0.1, 1, 10]

# The list of choice for hyperparameter kernel
kernels = ['linear', 'poly', 'rbf', 'sigmoid']

# The list of choice for hyperparameter gamma
gammas = ['auto', 1, 10, 100, 1000]

# The list of [score, setting]
# where score is the score of the classifier and setting a list of [C, kernel, gamma]
score_settings = []

# For each C
for C in Cs:
    # For each kernel
    for kernel in kernels:
        # For each gamma
        for gamma in gammas:
            # Declare the classifier with hyperparameter C, kernel, gamma, and random_state (0)
            # Implement me
            clf = SVC(C=C, kernel=kernel, gamma=gamma, random_state=0)

            # The pipeline, with StandardScaler and clf defined above
            # Implement me
            pipe_clf = Pipeline([('StandardScaler', StandardScaler()), ('SVC', clf)])

            # Get the score with respect to each fold (using cross_val_score with n_splits=10, shuffle=True, random_state=0)
            # Implement me
            scores = cross_val_score(estimator=pipe_clf,
                                     X=X,
                                     y=y,
                                     scoring='accuracy',
                                     cv=StratifiedKFold(n_splits=10,
                                                        shuffle=True,
                                                        random_state=0),
                                     n_jobs=-1)

            # Get the setting of C, kernel, and gamma
            setting = [C, kernel, gamma]

            # Get the average score (rounding to two decimal places)
            # Implement me            
            mean_score = round(np.mean(scores), 2)

            # Append [mean_score, setting] to score_settings
            # Implement me
            score_settings.append([mean_score, setting])
        
# Sort score_settings in descending order of score
# Implement me
score_settings = sorted(score_settings, key=lambda x: x[0], reverse=True)

# Print score_settings
print('The list of [score, setting] is:')
for score_setting in score_settings:
    print(score_setting)
print()

# Print the best setting
print('The best setting is:')
print('C:', score_settings[0][1][0])
print('kernel:', score_settings[0][1][1])
print('gamma:', score_settings[0][1][2])
```

# Discussion
As in the midterm, we use a nested for-loop for fine-tuning the hyperparameters of a model. While this shows us clearly what happens under the hood, it is by no means the best practice.

# The best practice for hyperparameter tuning
- Here we use the combination of [Pipeline](https://scikit-learn.org/stable/modules/generated/sklearn.pipeline.Pipeline.html) and [GridSearchCV](https://scikit-learn.org/stable/modules/generated/sklearn.model_selection.GridSearchCV.html) for hyperparameter tuning

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
from sklearn.model_selection import GridSearchCV

# Declare the classifier with random_state=0
# Implement me
clf = SVC(random_state=0)

# The pipeline, with StandardScaler and clf defined above
# Implement me
pipe_clf = Pipeline([('StandardScaler', StandardScaler()), ('clf', clf)])

# The parameter grid
param_grid = [{'clf__C': Cs,
               'clf__kernel': kernels,
               'clf__gamma': gammas}]

# GridSearchCV
gs = GridSearchCV(estimator=pipe_clf,
                  param_grid=param_grid,
                  scoring='accuracy',
                  n_jobs=-1,
                  cv=StratifiedKFold(n_splits=10,
                                     shuffle=True,
                                     random_state=0))

# Hyperparameter tuning and model selection
gs.fit(X, y)

# Get the cross-validation results
cv_results = pd.DataFrame.from_dict(gs.cv_results_).sort_values(by=['rank_test_score', 'std_test_score'])

# Write the cross-validation results to csv file
cv_results.to_csv(path_or_buf='cv_results.csv')

# Print the best settings
print('The best setting is:')
print(gs.best_params_)
```
